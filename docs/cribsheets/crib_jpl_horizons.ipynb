# Example of JPL Horizons in python

Installation: $ pip install astroquery

```python
from astroquery.jplhorizons import Horizons
```

```python
# Define the target (Venus:299) and observer (Earth:399)
obj = Horizons(id='299', location='500@399', epochs={'start':'2024-01-01', 'stop':'2024-01-05', 'step':'1d'})
```

```python
# Get ephemeris data
data = obj.ephemerides()
data.pprint(max_width=-1) 
```

```python
print(data['datetime_str', 'illumination'])  
```

```python
# Plot ephemeris data

import matplotlib.pyplot as plt
plt.plot(data['datetime_str'], data['illumination'], '+')
```
